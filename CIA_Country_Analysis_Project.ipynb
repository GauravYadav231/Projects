## Data Preparation and Model Discovery

Preparing data for Kmeans Clustering!

### Missing Data

```python
df.isnull().sum()
```

```python
df[df['Agriculture'].isnull()]['Country']
```

```python
# REMOVAL OF TINY ISLANDS
df[df['Agriculture'].isnull()] = df[df['Agriculture'].isnull()].fillna(0)
```

```python
df.isnull().sum()
```

```python
# https://stackoverflow.com/questions/19966018/pandas-filling-missing-values-by-mean-in-each-group
df['Climate'] = df['Climate'].fillna(df.groupby('Region')['Climate'].transform('mean'))
```

**Checking again on many elements are missing:**

```python
df.isnull().sum()
```

**Using the same tactic as we did with Climate missing values and fill in any missing Literacy % values with the mean Literacy % of the Region.**

```python
df[df['Literacy (%)'].isnull()]
```

```python
# https://stackoverflow.com/questions/19966018/pandas-filling-missing-values-by-mean-in-each-group
df['Literacy (%)'] = df['Literacy (%)'].fillna(df.groupby('Region')['Literacy (%)'].transform('mean'))
```

**Checking again on the remaining missing values:**

```python
df.isnull().sum()
```

**Only a few countries so for simplicity, we will drop these.**

```python
df = df.dropna()
```

## Data Feature Preparation

**Preparing the data for clustering. 
The Country column is still a unique identifier string, so it won't be useful for clustering, since its unique for each point, hence we will drop Country column.**

```python
X = df.drop("Country",axis=1)
```

```python
X = pd.get_dummies(X)
```

```python
X.head()
```

### Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
```

```python
scaled_X
```

### Creating and Fitting Kmeans Model




**Using a for loop to create and fit multiple KMeans models, testing from K=2-30 clusters.**

```python
from sklearn.cluster import KMeans
```

```python
ssd = []

for k in range(2,30):
    
    model = KMeans(n_clusters=k)
    
    
    model.fit(scaled_X)
    
    #Sum of squared distances of samples to their closest cluster center.
    ssd.append(model.inertia_)
```

```python
plt.plot(range(2,30),ssd,'o--')
plt.xlabel("K Value")
plt.ylabel(" Sum of Squared Distances")
```

```python
pd.Series(ssd).diff().plot(kind='bar')
```
